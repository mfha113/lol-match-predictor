```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('high_diamond_ranked_10min.csv')

#Check the first few rows
df.head()
```

```python
df.info()

df.describe().transpose()
```

```python
# Fill missing values with median (if any exist)
df.fillna(df.median(), inplace=True)
```

```python
# Check for duplicate rows
duplicates = df.duplicated().sum()
print(f"Duplicate rows: {duplicates}")

# Remove duplicates if needed
df.drop_duplicates(inplace=True)
```

```python
# Check the distribution of blueWins
sns.countplot(x='blueWins', data=df)
plt.title("Distribution of Match Outcomes")
plt.xlabel("Blue Team Wins (1) vs Loses (0)")
plt.ylabel("Number of Matches")
plt.show()
```

```python
# Compute the correlation matrix
corr_matrix = df.corr()

# Correlation heatmap
plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=False)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Check the highest correlations with blueWins
corr_target = corr_matrix["blueWins"].abs().sort_values(ascending=False)
print("Top Features correlated with match outcome:")
print(corr_target.head(10))
```

```python
# Visualization for important features from above
important_features = corr_target.index[1:6]

for feature in important_features:
    plt.figure(figsize=(8,4))
    sns.boxplot(x='blueWins', y=feature, data=df)
    plt.title(f'Feature: {feature}')
    plt.xlabel("Match Outcome (0: Loss, 1: Win)")
    plt.ylabel(feature)
    plt.show()
```

```python
# Separate blueWins
X = df.drop(columns='blueWins')
y = df['blueWins']

# Feature scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X), columns=X.columns)

X_scaled.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=28)

print(f'Training Set Size: {X_train.shape[0]}')
print(f'Test Set Size: {X_test.shape[0]}')
```

```python
# Save processed data
X_train.to_csv('features_train_scaled.csv', index=False)
X_test.to_csv('features_test_scaled.csv', index=False)
y_train.to_csv('target_train.csv', index=False)
y_test.to_csv('target_test.csv', index=False)
```
